```python
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter
import numpy as np
%matplotlib inline
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('png', 'pdf')
```

```python
data = np.loadtxt('V2227sed.dat')
# second column is Fnu in microJy, 4th column is lFl
wl = data[:,0]
logwl = np.log10(wl)
f2227 = data[:,1] 
f2227 = np.log10(f2227 / f2227[0])
e2227 = data[:,4]

data = np.loadtxt('V1959sed.dat')
f1959 = data[:,1]
f1959 = np.log10(f1959 / f1959[0])
e1959 = data[:,4]

data = np.loadtxt('V1999sed.dat')
f1999 = data[:,1] 
f1999 = np.log10(f1999 / f1999[0])
e1999 = data[:,4]

data = np.loadtxt('V2559sed.dat')
f2559 = data[:,1]
f2559 = np.log10(f2559 / f2559[0])
e2559 = data[:,4]
```

```python
# templates
labtemp = np.array([1.235,1.662,2.159,3.4,4.6,12,22])
zero = np.array([1594,1024,667,309.540,171.787,31.674,8.363])
# AU Mic M1 (Cotten & Song 2016)
magmic = np.array([5.436,4.831,4.529,4.449,4.008,4.313,4.140])
magmic = magmic * -0.4
fnumic = zero * 10**(magmic)
fnumic = np.log10(fnumic / fnumic[0])
# UscoCtio13 (Luhman & Mamajek) -- evolved
magevo = np.array([12.169,11.497,11.188,10.858,10.499,9.313,7.074])
magevo = magevo * -0.4
fnuevo = zero * 10**(magevo)
fnuevo = np.log10(fnuevo / fnuevo[0])
# UscoCtio33 (Luhman & Mamajek) -- full
magful = np.array([12.297,11.670,11.300,10.771,10.163,8.258,6.163])
magful = magful * -0.4
fnuful = zero * 10**(magful)
fnuful = np.log10(fnuful / fnuful[0])
```

```python
%matplotlib inline
plt.rc('font', family='serif')
plt.rc('xtick', labelsize=10)
plt.rc('ytick', labelsize=10)
fig = plt.figure(figsize=(10, 8))
```

```python
uppers_wl = []
uppers_fl =[]
wlnum = len(wl)
wlnumar = np.arange(wlnum)
for x in wlnumar:
  if (e2227[x] == 0.0) & (f2227[x] >-5.0):
    uppers_wl.append(wl[x])
    uppers_fl.append(f2227[x])
uppers_wl_np = np.array(uppers_wl)
uppers_fl_np = np.array(uppers_fl)

ax3 = plt.subplot(223)
plt.ylabel(r'$log F_\nu$',fontsize=15)
plt.xlabel('$\lambda (\mu m$)',fontsize=15)
plt.ylim(-1.0,0.5)
plt.yticks([-0.8,-0.4,0.0,0.4])
plt.text(1.2,0.3,'V2227',fontsize=15)
plt.errorbar(uppers_wl_np,uppers_fl_np,xerr=None,yerr=0.1,fmt='none',ecolor='k',capsize=3,uplims=True)
plt.semilogx(wl,f2227,'o', color='r')
plt.semilogx(labtemp,fnumic,'--',color='k')
plt.semilogx(labtemp,fnuevo,'-',color='k')
plt.semilogx(labtemp,fnuful,'-',color='k')
plt.xlim(1.0,30.0)
ax3.xaxis.set_major_formatter(ScalarFormatter())

uppers_wl = []
uppers_fl =[]
wlnum = len(wl)
wlnumar = np.arange(wlnum)
for x in wlnumar:
  if (e1959[x] == 0.0) & (f1959[x] >-5.0):
    uppers_wl.append(wl[x])
    uppers_fl.append(f1959[x])
uppers_wl_np = np.array(uppers_wl)
uppers_fl_np = np.array(uppers_fl)

ax1 = plt.subplot(221,sharex=ax3)
plt.ylabel(r'$log F_\nu$',fontsize=15)
plt.ylim(-1.0,0.5)
plt.xlim(1.0,30.0)
plt.yticks([-0.8,-0.4,0.0,0.4])
plt.setp(ax1.get_xticklabels(), visible=False)
plt.text(1.2,0.3,'V1959',fontsize=15)
plt.errorbar(uppers_wl_np,uppers_fl_np,xerr=None,yerr=0.1,fmt='none',ecolor='k',capsize=3,uplims=True)
plt.semilogx(wl,f1959,'o', color='r')
plt.semilogx(labtemp,fnumic,'--',color='k')
plt.semilogx(labtemp,fnuevo,'-',color='k')
plt.semilogx(labtemp,fnuful,'-',color='k')
plt.xlim(1.0,30.0)
ax1.xaxis.set_major_formatter(ScalarFormatter())

uppers_wl = []
uppers_fl =[]
wlnum = len(wl)
wlnumar = np.arange(wlnum)
for x in wlnumar:
  if (e2559[x] == 0.0) & (f2559[x] >-5.0):
    uppers_wl.append(wl[x])
    uppers_fl.append(f2559[x])
uppers_wl_np = np.array(uppers_wl)
uppers_fl_np = np.array(uppers_fl)

ax4 = plt.subplot(224,sharey=ax3)
plt.xlabel('$\lambda (\mu m$)',fontsize=15)
plt.ylim(-1.0,0.5)
plt.xlim(1.0,30.0)
plt.xticks([2.0,5.0,10.0,20.0])
plt.setp(ax4.get_yticklabels(), visible=False)
plt.text(1.2,0.3,'V2559',fontsize=15)
plt.errorbar(uppers_wl_np,uppers_fl_np,xerr=None,yerr=0.1,fmt='none',ecolor='k',capsize=3,uplims=True)
plt.semilogx(wl,f2559,'o', color='r')
plt.semilogx(labtemp,fnumic,'--',color='k')
plt.semilogx(labtemp,fnuevo,'-',color='k')
plt.semilogx(labtemp,fnuful,'-',color='k')
plt.xlim(1.0,30.0)
ax4.xaxis.set_major_formatter(ScalarFormatter())

uppers_wl = []
uppers_fl =[]
wlnum = len(wl)
wlnumar = np.arange(wlnum)
for x in wlnumar:
  if (e1999[x] == 0.0) & (f1999[x] >-5.0):
    uppers_wl.append(wl[x])
    uppers_fl.append(f1999[x])
uppers_wl_np = np.array(uppers_wl)
uppers_fl_np = np.array(uppers_fl)

ax2 = plt.subplot(222,sharey=ax1,sharex=ax4)
plt.ylim(-1.0,0.5)
plt.xlim(1.0,30.0)
plt.setp(ax2.get_xticklabels(), visible=False)
plt.setp(ax2.get_yticklabels(), visible=False)
plt.text(1.2,0.3,'V1999',fontsize=15)
plt.errorbar(uppers_wl_np,uppers_fl_np,xerr=None,yerr=0.1,fmt='none',ecolor='k',capsize=3,uplims=True)
plt.semilogx(wl,f1999,'o', color='r')
plt.semilogx(labtemp,fnumic,'--',color='k',label='AUMic')
plt.semilogx(labtemp,fnuevo,'-',color='k',label='XXX')
plt.semilogx(labtemp,fnuful,'-',color='k',label='XXX')
plt.xlim(1.0,30.0)
ax2.xaxis.set_major_formatter(ScalarFormatter())

plt.tight_layout(h_pad=0.0)
plt.show()
```
