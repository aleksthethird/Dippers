```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('png', 'pdf')
```

```python
data = np.loadtxt('ts7895.txt') 
V1959t = data[:,0]
V1959m = data[:,1]

data = np.loadtxt('ts7912.txt') 
V1999t = data[:,0]
V1999m = data[:,1]

data = np.loadtxt('ts1791.txt') 
V2227t = data[:,0]
V2227m = data[:,1]

data = np.loadtxt('ts10860.txt') 
V2559t = data[:,0]
V2559m = data[:,1]
```

```python
%matplotlib inline
plt.rc('font', family='serif')
plt.rc('xtick', labelsize=15)
plt.rc('ytick', labelsize=15)
fig = plt.figure(figsize=(10, 8))
```

```python
fig = plt.figure(figsize=(10, 8))
plt.rc('font', family='serif')
plt.rc('xtick', labelsize=15)
plt.rc('ytick', labelsize=15)

ax5 = plt.subplot(224)
plt.plot(V2559t, V2559m, 'o', color='k')
#plt.ylabel('I-band',fontsize=20)
plt.xlabel('Time (d)',fontsize=20)
plt.ylim(0.8,-0.4)
plt.setp(ax5.get_xticklabels(), fontsize=15)
plt.yticks([0.6,0.3,0.0,-0.3])
plt.text(0.1,0.7,'V2559',fontsize=20)

ax3=plt.subplot(223)
plt.ylim(0.15,-0.12)
plt.plot(V2227t, V2227m, 'o', color='k')
plt.ylabel('I-band',fontsize=20)
plt.xlabel('Time (d)',fontsize=20)
plt.setp(ax3.get_xticklabels(), visible=True)
plt.yticks([0.1,0.0,-0.1])
plt.text(0.1,0.14,'V2227',fontsize=20)

ax1 = plt.subplot(221,sharex=ax3)
plt.plot(V1959t, V1959m, 'o', color='k')
plt.ylabel('I-band',fontsize=20)
plt.ylim(0.15,-0.05)
plt.setp(ax1.get_xticklabels(), visible=False)
plt.yticks([0.1,0.0])
plt.text(0.3,0.14,'V1959',fontsize=20)

ax2 = plt.subplot(222,sharex=ax5)
plt.plot(V1999t, V1999m, 'o', color='k')
#plt.ylabel('I-band',fontsize=20)
plt.ylim(0.5,-0.25)
plt.setp(ax2.get_xticklabels(), visible=False)
plt.yticks([0.3,0.1,-0.1])
plt.text(0.1,0.45,'V1999',fontsize=20)

plt.tight_layout(h_pad=0.0)
plt.show()
```

```python
data = np.loadtxt('se2005wise.cat')
j = data[:,13]
w1 = data[:,9]
w2 = data[:,11]
w1_w2 = w1 - w2
```

```python
# this is straight from WISE on Vizier
W1var = np.array([14.29,12.77,13.68,12.23])
W2var = np.array([13.81,12.36,13.47,12.05])
Jvar = np.array([15.73,14.13,14.76,13.34])
W1_W2var = W1var - W2var
```

```python
%matplotlib inline
plt.rc('font', family='serif')
plt.rc('xtick', labelsize=15)
plt.rc('ytick', labelsize=15)
fig = plt.figure(figsize=(8, 6))
plt.xlim(0.0,0.65)
plt.ylim(18,10)
plt.plot(w1_w2,j,'ko')
plt.title('WISE colours in SE2005 sample',fontsize=20)
plt.xlabel('W1-W2',fontsize=20)
plt.ylabel('J (mag)',fontsize=20)
plt.plot([0.4,0.4],[10,18],'k--')
plt.plot(W1_W2var,Jvar,'r*',markersize=20)
plt.show()
```
