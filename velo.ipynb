```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('png', 'pdf')
```

```python
velo_kms = np.array([30,30,200,4])
# as derived from the typical transit time
Mstar = 0.4
orbit = 6.672e-11 * Mstar * 2e30 / (velo_kms*1e3)**2
orbit_AU = orbit / 150e9
print(orbit_AU)
delta_r = np.array([0.23,0.81,0.36,1.44])
log_orbit_AU = np.log10(orbit)
log_vel_kms = np.log10(velo_kms)
```

```python
orbit_AU_test = np.array([0.1,1,10])
orbit_test = orbit_AU_test * 150e9
velo_kms_test = np.sqrt( 6.672e-11 * Mstar * 2e30 / orbit_test) / 1000.0
log_velo_kms_test = np.log10(velo_kms_test)
print(log_velo_kms_test)
```

```python
disk = [500,200,100,200]
# 2 for full/evolved, 2 for evolved, 0 for debris
```

```python
%matplotlib inline
plt.rc('font', family='serif')
plt.rc('xtick', labelsize=15)
plt.rc('ytick', labelsize=15)
fig = plt.figure(figsize=(8, 6))
plt.ylim(0.1,1.8)
plt.xlim(0.2,2.5)
plt.text(log_velo_kms_test[0],1.6,'0.1AU',fontsize=15,color='g')
plt.text(log_velo_kms_test[1],1.6,'1AU',fontsize=15,color='g')
plt.text(log_velo_kms_test[2],1.6,'10AU',fontsize=15,color='g')
plt.text(2.16,0.45,'V2227',fontsize=15)
plt.text(1.34,0.36,'V1959',fontsize=15)
plt.text(1.34,0.90,'V1999',fontsize=15)
plt.text(0.48,1.3,'V2559',fontsize=15)
plt.xlabel('Log(v (km/s))',fontsize=20)
plt.ylabel('LCO i-mag amplitude (mag)',fontsize=20)
plt.scatter(log_vel_kms,delta_r,s=disk)
```
