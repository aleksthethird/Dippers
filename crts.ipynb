```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('png', 'pdf')
```

```python
data = np.loadtxt('v1959css.csv',delimiter=',') 
V1959t = data[:,5]
V1959m = data[:,1]

data = np.loadtxt('v1999css.csv',delimiter=',') 
V1999t = data[:,5]
V1999m = data[:,1]

data = np.loadtxt('v2227css.csv',delimiter=',') 
V2227t = data[:,5]
V2227m = data[:,1]

data = np.loadtxt('v2559css.csv',delimiter=',') 
V2559t = data[:,5]
V2559m = data[:,1]
```

```python
plt.rc('font', family='serif')
plt.rc('xtick', labelsize='15')
plt.rc('ytick', labelsize='15')
fig = plt.figure(figsize=(10, 8))

ax4 = plt.subplot(224)
plt.plot(V2559t, V2559m,'o',color='k')
plt.xlabel('MJD (d)',fontsize=20)
plt.ylim(18.5,15.5)
plt.setp(ax4.get_xticklabels(), fontsize=15)
plt.xticks([54000,55000,56000])
plt.text(55500,18.3,'V2559',fontsize=20)

ax3=plt.subplot(223)
plt.ylim(18.5,15.5)
plt.plot(V2227t, V2227m,'o',color='k')
plt.ylabel('V-band',fontsize=20)
plt.xlabel('MJD (d)',fontsize=20)
plt.setp(ax3.get_xticklabels(), visible=True)
plt.xticks([54000,55000,56000])
plt.text(55500,18.3,'V2227',fontsize=20)

ax1 = plt.subplot(221,sharex=ax3)
plt.plot(V1959t, V1959m,'o',color='k')
plt.ylabel('V-band',fontsize=20)
plt.ylim(19.5,16.5)
#plt.title('Lightcurves from CRTS')
plt.setp(ax1.get_xticklabels(), visible=False)
plt.xticks([54000,55000,56000])
plt.text(55500,19.3,'V1959',fontsize=20)

ax2 = plt.subplot(222,sharex=ax4)
plt.plot(V1999t, V1999m,'o',color='k')
plt.ylim(21.0,18.0)
plt.setp(ax2.get_xticklabels(), visible=False)
plt.text(55500,20.8,'V1999',fontsize=20)
plt.xticks([54000,55000,56000])
plt.tight_layout(h_pad=0.0)
plt.show()
```
